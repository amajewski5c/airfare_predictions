```python
sdd = sdf[s_cols]

sdd['Yfare'] = sdd['fare'].to_numpy()/sdd['passengers'].to_numpy()

sdd = sdd[(sdd.carrier_1 == sdd.carrier_2) ].copy()
sdd
```

```python
sdd.cabin_class.value_counts()
```

```python
def clean_class(val):
    return val.split(' ')[0].replace('_','').replace('mEconomy','m').replace('premiumeconomy', 'premium').lower()

sdd['cleanclass'] = [ clean_class(x) for x in sdd['cabin_class'] ]
sdd.cleanclass.value_counts()
```

```python
sdd = sdd[~sdd.cleanclass.isin(['coach','mixed'])]
sdd.cleanclass.value_counts()
```
